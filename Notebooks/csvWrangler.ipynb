```python
import csv
from datetime import datetime
import datetime as dt

#import productivity per hour data

#bring all data in from csv into array
data = []
with open('productivityPerHour.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:  
        data.append(row)
startYear = data[0][1]
data = data[1:]
#print(startYear)
#print(data)          
```

```python
#Setup array for data once its in correct form
dataBase = [] #[country,year,person or hour, rate, comment] # 0 = hour, 1 = person


#ingest all data and insert in correct form
for item in data:
    #print(item)
    year = int(startYear)
    country = item[0]
    for i in range(1,len(item)-1,2):
        rate = item[i]
        if(rate == ':'):
            rate = ''
        else:
            rate = float(rate)
        comment = item[i+1]
        if(len(comment) > 0):  #comments are of the form '(p)' so this makes comment go to just 'p'
            comment = comment[1]
        dataBase.append([country,year,0,rate,comment])
        year = year + 1
```

```python
# do it again for per person
#bring all data in from csv into array
data = []
with open('productivityPerPerson.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:  
        data.append(row)
startYear = data[0][1]
data = data[1:]
#print(startYear)
#print(data)     
```

```python
#ingest all data and insert in correct form
for item in data:
    #print(item)
    year = int(startYear)
    country = item[0]
    for i in range(1,len(item)-1,2):
        rate = item[i]
        if(rate == ':'):
            rate = ''
        else:
            rate = float(rate)
        comment = item[i+1]
        if(len(comment) > 0):  #comments are of the form '(p)' so this makes comment go to just 'p'
            comment = comment[1]
        dataBase.append([country,year,1,rate,comment])
        year = year + 1
```

```python
#check database is looking how you expect
for item in dataBase:
    print(item)
```

```python
#output csv of all productivity data
with open('productivityData.csv', mode='w' , newline='') as outputFile:
    outputWriter = csv.writer(outputFile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for d in dataBase:
        outputWriter.writerow(d)
        #print(d)
```
