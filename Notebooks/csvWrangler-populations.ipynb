```python
import csv
from datetime import datetime
import datetime as dt

#import productivity per hour data

#bring all data in from csv into array
data = []
with open('population.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:  
        data.append(row)

print(data)          
```

```python
#Setup array for data once its in correct form
dataBase = [] #[country,year,sex, population] # 0 = both, 1 = male, 2 = female

sex = 0
country = ''
startYear = 1900

stageCount = 0
#ingest all data and insert in correct form
for item in data:
    #print(item)
    #Handle Gender
    if(item[0] == 'SEX'):
        stageCount = 1
        if(item[1] == 'Total'):
            sex = 0
        elif(item[1] == 'Males'):
            sex = 1
        elif(item[1] == 'Females'):
            sex = 2
        else:
            print("Gender Error" + item[1])         
    #Handle all others
    else:
        stageCount = stageCount + 1
        #still metadata so skip
        if(stageCount == 5):
            startYear = int(item[1])
        elif(stageCount > 5):
            #check its not blank
            if(item != ['', '', '', '', '', '', '', '', '', '', '']):
                country = item[0]
                year = startYear
                for i in range(1,len(item)):
                    population = item[i]
                    if(population == ':'):
                        population = ''
                    else:
                        population = int(population)
                    dataBase.append([country,year,sex,population])
                    year = year + 1
            #finished with that set reset counter
            else:
                stageCount = -10
```

```python
#check database is looking how you expect
for item in dataBase:
    print(item)
```

```python
#output csv of all productivity data
with open('allPopulationData.csv', mode='w' , newline='') as outputFile:
    outputWriter = csv.writer(outputFile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for d in dataBase:
        outputWriter.writerow(d)
        #print(d)
```
